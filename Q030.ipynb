```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q030[SQ001] - Q030[SQ007]

agree_columns = ['Q030[SQ001]','Q030[SQ002]','Q030[SQ003]','Q030[SQ004]','Q030[SQ005]','Q030[SQ006]','Q030[SQ007]']
agree_options = [u"Informazioni di carattere generale sul making e la comunità dei maker",
                 u"Conoscenze tecniche e tecnologiche sui processi di progettazione e produzione",
                 u"Files e dati relativi a progetti esistenti",
                 u"Sviluppo di nuovi progetti",
                 u"Tecnologie, strumenti, macchine e attrezzature",
                 u"Spazi e ambienti di lavoro",
                 u"Organizzazione di nuove iniziative"]
agree = data[agree_columns]
agree_types = [u"Moltissimo",u"Molto",u"Abbastanza",u"Poco",u"Niente",u"Non so / Non risponde"]
```

```python
[u"Moltissimo",u"Molto",u"Abbastanza",u"Poco",u"Niente",u"Non so / Non risponde"]

# Gather data
agree_b = {}
for k,i in enumerate(agree_columns):
    agree_b[k] = agree[i].value_counts()
    print "Data:",agree_options[k]
    print agree_b[k]
    print
    print "Data %:",agree_options[k]
    print agree[i].value_counts(normalize=True)*100
    print

moltissimo = []
molto = []
abbastanza = []
poco = []
niente = []
nonso = []

for k,i in enumerate(agree_columns):
    agree_presents = agree_b[k].index.tolist()
    # Check for empty values, and put a 0 instead
    if u"Moltissimo" not in agree_presents:
        moltissimo.append(0)
    if u"Molto" not in agree_presents:
        molto.append(0)
    if u"Abbastanza" not in agree_presents:
        abbastanza.append(0)
    if u"Poco" not in agree_presents:
        poco.append(0)
    if u"Niente" not in agree_presents:
        niente.append(0)
    if u"Non so / Non risponde" not in agree_presents:
        nonso.append(0)
        
    for j in agree_presents:
        if j == u"Moltissimo":
            moltissimo.append(agree_b[k].ix[u"Moltissimo"])
        elif j == u"Molto":
            molto.append(agree_b[k].ix[u"Molto"])
        elif j == u"Abbastanza":
            abbastanza.append(agree_b[k].ix[u"Abbastanza"])
        elif j == u"Poco":
            poco.append(agree_b[k].ix[u"Poco"])
        elif j == u"Niente":
            niente.append(agree_b[k].ix[u"Niente"])
        elif j == u"Non so / Non risponde":
            nonso.append(agree_b[k].ix[u"Non so / Non risponde"])
            
# Plot the data
plt.figure(figsize=(12,6))
plt.xlabel('Frasi')
plt.ylabel('Persone')
plt.title("Quanto sono importanti per te lo scambio, la condivisione e la collaborazione rispetto a...")
plt.xticks(range(len(agree_options)),agree_options,rotation=90)
ind = np.arange(len(agree_columns))       # the x locations for the groups
width = 0.15                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 6) # Set color palette
rect1 = plt.bar(ind,moltissimo,width,color=my_colors[0],align='center') # Plot moltissimo
rect2 = plt.bar(ind+width,molto,width,color=my_colors[1],align='center') # Plot molto 
rect3 = plt.bar(ind+width*2,abbastanza,width,color=my_colors[2],align='center') # Plot abbastanza
rect4 = plt.bar(ind+width*3,poco,width,color=my_colors[3],align='center') # Plot poco
rect5 = plt.bar(ind+width*4,niente,width,color=my_colors[4],align='center') # Plot niente
rect6 = plt.bar(ind+width*5,nonso,width,color=my_colors[5],align='center') # Plot non so
plt.legend( (rect1, rect2, rect3, rect4, rect5, rect6), agree_types )
plt.savefig("pdf/Q030-MakerAccordo01.pdf")
```
