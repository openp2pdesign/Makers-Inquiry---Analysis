# Q047 - Offri servizi collegati al prodotto?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q047[SQ001] - Q047[SQ004] - Q047[other]

services_columns = ['Q047[SQ001]','Q047[SQ002]','Q047[SQ003]','Q047[SQ004]']
services_options = ['Consegna a domicilio',
                  u'Certificato di qualitá',
                  'Assistenza postvendita (riparazione, sostituzione)',
                  'Garanzia']
services = data[services_columns]
services.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
services_other = data['Q047[other]'].str.lower().value_counts()
```

```python
services[0:4]
```

```python
# Gather data
services_b = {}
for k,i in enumerate(services_columns):
    services_b[k] = services[i].value_counts()
    print "Data:",services_options[k]
    print services_b[k]
    print
    print "Data %:",services_options[k]
    print services[i].value_counts(normalize=True)*100
    print
    
yes = []
no = []

# Get data for plotting, and check that we have all the values for Yes and No
for k,i in enumerate(services_columns):
    if len(services_b[k].index) == 2:
        if services_b[k].index[0] == "Si":
            yes.append(services_b[k].iloc[0])
            no.append(services_b[k].iloc[1])
        elif services_b[k].index[1] == "Si":
            yes.append(services_b[k].iloc[1])
            no.append(services_b[k].iloc[0])
    elif len(services_b[k].index) == 1:
        if services_b[k].index[0] == "Si":
            yes.append(services_b[k].iloc[0])
            no.append(0)
        elif services_b[k].index[0] == "No":
            yes.append(0)
            no.append(services_b[k].iloc[0])
    else:
        yes.append(0)
        no.append(0)

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Servizi offerti')
plt.ylabel('Persone')
plt.title('Offri anche un servizio?')
plt.xticks(range(len(services_options)),services_options,rotation=90)
ind = np.arange(len(services_columns))   # the x locations for the groups
width = 0.35                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 2) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[1],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[0],align='center') # Plot No 
plt.legend( (rect1, rect2), ('Si', 'No') )
plt.savefig("pdf/Q047-MakerServiziOfferti01.pdf")
```
