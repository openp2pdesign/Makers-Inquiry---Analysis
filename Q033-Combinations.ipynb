```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn
import shutil

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html

# Change the font
matplotlib.rcParams.update({'font.family': 'Source Sans Pro'})
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q033[SQ001] - Q055[SQ010] - Q055[other]

activities_columns = ['Q033[SQ001]','Q033[SQ002]','Q033[SQ003]','Q033[SQ004]']
activities_options = [u'Seguo alcune delle attivitá organizzate',
                    'Sviluppo i miei progetti',
                    'Sviluppo progetti con altre persone che frequentano lo spazio',
                    'Organizzo iniziative ed eventi con altre persone che frequentano lo spazio'
                    ]
activities = data[activities_columns]
activities.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
activities_other = data['Q055[other]'].str.lower().value_counts()
```

## Combinations...

```python
# Create all the possible combinations from the main options
# See http://stackoverflow.com/questions/17176887/python-get-all-permutation-of-a-list-w-o-repetitions

import itertools 

all_combinations = {}
all_combinations_columns = []

for i in range(1, len(activities_columns)+1):
    comb = list(itertools.combinations(activities_columns, i))
    for k in comb:
        # Each combination
        all_combinations[k] = {}
        all_combinations[k]["col_list"] = list(k)
        # Build the string and boolean list of each combination
        comb_list = []
        comb_bool_list = []
        # Put default False value
        for l in activities_columns:
            comb_bool_list.append(False)
        for j in k:
            pos = activities_columns.index(j) # Get position
            comb_list.append(activities_options[pos])
            comb_bool_list[pos] = True
        all_combinations[k]["list"] = comb_list
        all_combinations[k]["bool_list"] = comb_bool_list
        all_combinations[k]["str"] = ", ".join(comb_list)
```

```python
# Check which combinations correspond each row
str_values = []
for i in activities.index:
    current_bool_list = list(activities.ix[i].isin(["Si"]))
    for i in all_combinations:
        if current_bool_list == all_combinations[i]["bool_list"]:
            str_values.append(all_combinations[i]["str"])
```

```python
# Add combination column
activities["Combination"] = pd.Series(str_values)
```

```python
# Gather data
resulting_combinations = activities["Combination"].value_counts(dropna=False)
resulting_combinations_percentage = activities["Combination"].value_counts(normalize=True,dropna=False)*100
```

```python
# Plotting the first 10 values of the most popular combinations
resulting_combinations[0:10].plot(kind='bar',figsize=(20,10),rot=90)
plt.title(u"Che attivitá svolgi all'interno di questo luogo? Combinazioni", fontsize=18, y=1.02)
plt.ylabel("Persone", fontsize=16)
plt.xlabel("Combinazioni", fontsize=16)
plt.savefig("svg/Q033-Combinazioni.svg")
plt.savefig("png/Q033-Combinazioni.png")
plt.savefig("pdf/Q033-Combinazioni.pdf")
```
