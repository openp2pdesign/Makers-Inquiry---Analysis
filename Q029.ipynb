```python
# Range: Q029[SQ001] - Q029[SQ005]

agree_columns = ['Q029[SQ001]','Q029[SQ002]','Q029[SQ003]','Q029[SQ004]','Q029[SQ005]']
agree_options = ["I makers hanno un ruolo sociale","I makers hanno un ruolo politico",
                 "Le istituzioni non capiscono il ruolo del maker","Makers e hackers sono due figure distinte",
                 "I makers costituiscono un nuovo movimento"]
agree = data[agree_columns]
agree_types = [u"Completamente d’accordo",u"Parzialmente d’accordo",u"In disaccordo",u"Totalmente In disaccordo",u"Non so Non rispondo"]
```

```python
agree
```

```python
# Gather data
agree_b = {}
for k,i in enumerate(agree_columns):
    agree_b[k] = agree[i].value_counts()
    print "Data:",agree_options[k]
    print agree_b[k]
    print
    print "Data %:",agree_options[k]
    print agree[i].value_counts(normalize=True)*100
    print

completamente = []
parzialmente = []
disaccordo = []
totalmente = []
nonso = []

for k,i in enumerate(agree_columns):
    agree_presents = agree_b[k].index.tolist()
    # Check for empty values, and put a 0 instead
    if u"Completamente d\u2019accordo" not in agree_presents:
        completamente.append(0)
    if u"Parzialmente d\u2019accordo" not in agree_presents:
        parzialmente.append(0)
    if u"In disaccordo" not in agree_presents:
        disaccordo.append(0)
    if u"Totalmente  In disaccordo" not in agree_presents:
        totalmente.append(0)
    if u"Non so Non rispondo" not in agree_presents:
        nonso.append(0)
        
    for j in agree_presents:
        if j == u"Completamente d\u2019accordo":
            completamente.append(agree_b[k].ix[u"Completamente d\u2019accordo"])
        elif j == u"Parzialmente d\u2019accordo":
            parzialmente.append(agree_b[k].ix[u"Parzialmente d\u2019accordo"])
        elif j == u"In disaccordo":
            disaccordo.append(agree_b[k].ix[u"In disaccordo"])
        elif j == u"Totalmente  In disaccordo":
            totalmente.append(agree_b[k].ix[u"Totalmente  In disaccordo"])
        elif j == u"Non so Non rispondo":
            nonso.append(agree_b[k].ix[u"Non so Non rispondo"])
            
# Plot the data
plt.figure(figsize=(12,6))
plt.xlabel('Frasi')
plt.ylabel('Persone')
plt.title("Maker: sei d'accordo?")
plt.xticks(range(len(agree_options)),agree_options,rotation=90)
ind = np.arange(len(agree_columns))       # the x locations for the groups
width = 0.15                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 6) # Set color palette
rect1 = plt.bar(ind,completamente,width,color=my_colors[0],align='center') # Plot completamente
rect2 = plt.bar(ind+width,parzialmente,width,color=my_colors[1],align='center') # Plot parzialmente 
rect3 = plt.bar(ind+width*2,disaccordo,width,color=my_colors[2],align='center') # Plot disaccordo
rect4 = plt.bar(ind+width*3,totalmente,width,color=my_colors[3],align='center') # Plot totalmente
rect5 = plt.bar(ind+width*4,nonso,width,color=my_colors[4],align='center') # Plot non so
plt.legend( (rect1, rect2, rect3, rect4, rect5), agree_types )
plt.savefig("pdf/Q029-MakerAccordo01.pdf")
```
