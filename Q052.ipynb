# Q052 - Quali tecnologie hai acquistato negli ultimi 2 anni?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q052[SQ001] - Q052[SQ011] - Q052[other]

business_columns = ['Q052[SQ001]','Q052[SQ002]','Q052[SQ003]','Q052[SQ004]',
                    'Q052[SQ005]','Q052[SQ006]','Q052[SQ007]','Q052[SQ008]',
                    'Q052[SQ009]','Q052[SQ010]','Q052[SQ011]'
                    ]
business_options = ['Nessuna',
                  '3D Printer (fused filament fabrication)',
                  '3D Printer STL',
                  'Precision Desktop Milling Machine',
                  'Fresa CNC grande',
                  'Tornio CNC grande',
                  'Altri CNC di piccole dimensioni',
                  'Altri CNC di grandi dimensioni',
                  'Vinyl Cutter',
                  'Laser Cutter',
                  'Arduino, Raspberry PI, altri componenti elettronici e strumentazione per lavorazioni di elettronica'
                  ]
business = data[business_columns]
business.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
business_other = data['Q052[other]'].str.lower().value_counts()
```

```python
business[0:4]
```

```python
# Gather data
business_b = {}
for k,i in enumerate(business_columns):
    business_b[k] = business[i].value_counts()
    print "Data:",business_options[k]
    print business_b[k]
    print
    print "Data %:",business_options[k]
    print business[i].value_counts(normalize=True)*100
    print
    
yes = []
no = []

# Get data for plotting, and check that we have all the values for Yes and No
for k,i in enumerate(business_columns):
    if len(business_b[k].index) == 2:
        if business_b[k].index[0] == "Si":
            yes.append(business_b[k].iloc[0])
            no.append(business_b[k].iloc[1])
        elif business_b[k].index[1] == "Si":
            yes.append(business_b[k].iloc[1])
            no.append(business_b[k].iloc[0])
    elif len(business_b[k].index) == 1:
        if business_b[k].index[0] == "Si":
            yes.append(business_b[k].iloc[0])
            no.append(0)
        elif business_b[k].index[0] == "No":
            yes.append(0)
            no.append(business_b[k].iloc[0])
    else:
        yes.append(0)
        no.append(0)

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Tecnologie acquisite')
plt.ylabel('Persone')
plt.title(u'Quali tecnologie hai acquistato negli ultimi 2 anni?')
plt.xticks(range(len(business_options)),business_options,rotation=90)
ind = np.arange(len(business_columns))   # the x locations for the groups
width = 0.35                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 2) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[1],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[0],align='center') # Plot No 
plt.legend( (rect1, rect2), ('Si', 'No') )
plt.savefig("pdf/Q052-MakerTecnologieAcquisite01.pdf")
```
