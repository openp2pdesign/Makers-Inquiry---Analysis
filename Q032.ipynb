```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Get the distribution of study titles
lab= data["Q032"].value_counts(dropna=False)
print "Data:"
print lab
print ""
print "Data %:"
print data["Q032"].value_counts(normalize=True,dropna=False) * 100

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Tipi di laboratori frequentati')
plt.ylabel('Persone')
my_colors = seaborn.color_palette("husl", len(lab)) # Set color palette
lab.plot(kind="bar",title="Maker: Tipi di laboratori frequentati",color=my_colors)
plt.savefig("pdf/Q032-MakerTipiDiLaboratori01.pdf") # Save pdf
```
