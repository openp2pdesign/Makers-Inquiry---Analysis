```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html

# Change the font
matplotlib.rcParams.update({'font.family': 'Source Sans Pro'})
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q039[SQ001] - Q039[SQ005] - Q039[other]

approach_columns = ['Q039[SQ001]','Q039[SQ002]','Q039[SQ003]','Q039[SQ004]','Q039[SQ005]']
approach_options = ['Creo da zero i miei progetti',
                  'Utilizzo e modifico progetti esistenti (open design)',
                  'Sviluppo progetti in collaborazione con una community (codesign)',
                  'Utilizzo applicazioni che generano modelli (design generativo)',
                  'Sviluppo il progetto mentre lo realizzo materialmente (tinkering)']
approach = data[approach_columns]
approach.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
approach_other = data['Q039[other]'].str.lower().value_counts()
```

```python
#approach[0:4]
```

## Combinations...

```python
# Create all the possible combinations from the main options
# See http://stackoverflow.com/questions/17176887/python-get-all-permutation-of-a-list-w-o-repetitions

import itertools 

all_combinations = {}
all_combinations_columns = []

for i in range(1, len(approach_columns)+1):
    comb = list(itertools.combinations(approach_columns, i))
    for k in comb:
        # Each combination
        all_combinations[k] = {}
        all_combinations[k]["col_list"] = list(k)
        # Build the string and boolean list of each combination
        comb_list = []
        comb_bool_list = []
        # Put default False value
        for l in approach_columns:
            comb_bool_list.append(False)
        for j in k:
            pos = approach_columns.index(j) # Get position
            comb_list.append(approach_options[pos])
            comb_bool_list[pos] = True
        all_combinations[k]["list"] = comb_list
        all_combinations[k]["bool_list"] = comb_bool_list
        all_combinations[k]["str"] = ", ".join(comb_list)
```

```python
# Check which combinations correspond each row
str_values = []
for i in approach.index:
    current_bool_list = list(approach.ix[i].isin(["Si"]))
    for i in all_combinations:
        if current_bool_list == all_combinations[i]["bool_list"]:
            str_values.append(all_combinations[i]["str"])
```

```python
# Add combination column
approach["Combination"] = pd.Series(str_values)
```

```python
# Gather data
resulting_combinations = approach["Combination"].value_counts()
resulting_combinations_percentage = approach["Combination"].value_counts(normalize=True)*100
```

```python
# Plotting the first 10 values of the most popular combinations
resulting_combinations[0:10].plot(kind='bar',figsize=(20,10),rot=90)
plt.title(u"Qual é il tuo approccio alla progettazione? Combinazioni", fontsize=18, y=1.02)
plt.ylabel("Persone", fontsize=16)
plt.xlabel("Combinazioni", fontsize=16)
plt.savefig("svg/Q039-Combinazioni.svg")
plt.savefig("png/Q039-Combinazioni.png")
plt.savefig("pdf/Q039-Combinazioni.pdf")
```
