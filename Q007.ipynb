# Q007 - Qual è il tuo titolo di studio?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html

# Change the font
matplotlib.rcParams.update({'font.family': 'Source Sans Pro'})
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Get the distribution of study titles
study = data["Q007"].value_counts(dropna=False)
print "Data:"
print study
print ""
print "Data %:"
print data["Q007"].value_counts(normalize=True,dropna=False) * 100

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Titoli di studio', fontsize=16)
plt.ylabel('Persone', fontsize=16)
plt.title(u"Qual è il tuo titolo di studio?", fontsize=18, y=1.02)
my_colors = seaborn.color_palette("husl", len(study)) # Set color palette
study.plot(kind="bar",color=my_colors)
plt.savefig("svg/Q007-MakerTitoliDiStudio01.svg")
plt.savefig("png/Q007-MakerTitoliDiStudio01.png")
plt.savefig("pdf/Q007-MakerTitoliDiStudio01.pdf")
```

```python
# Get "other" data
study_other = data["Q007[other]"].str.lower().value_counts()
print "Data:"
print study_other
print ""
print "Data %:"
print data["Q007[other]"].str.lower().value_counts(normalize=True) * 100

# Plot bar
plt.figure(figsize=(8,6))
plt.title(u'Qual è il tuo titolo di studio? Altro', fontsize=18, y=1.02)
plt.xticks(range(len(study_other.index)),study_other.index,rotation=90)
plt.xlabel('Titolo', fontsize=16)
plt.ylabel('Persone', fontsize=16)
ind = np.arange(len(study_other))   # the x locations for the groups
width = 0.35                       # the width of the bars

my_colors = seaborn.color_palette("husl", len(study_other)) # Set color palette
rect1 = plt.bar(ind,study_other,width,color=my_colors,align='center')
plt.savefig("svg/Q007-MakerTitoliDiStudio02.svg")
plt.savefig("png/Q007-MakerTitoliDiStudio02.png")
plt.savefig("pdf/Q007-MakerTitoliDiStudio02.pdf")
```

```python
# Filter and integrate "other" data with data manually
study["Laurea specialistica (Master)"] += 7
study["Diploma scuola superiore (Upper secondary)"] += 2
study["Diploma post-laurea (Post-Master)"] += 1
study = study.drop("Altro")
study
```

```python
# Plot the updated data
plt.figure(figsize=(8,6))
plt.xlabel(u'Titoli di studio', fontsize=16)
plt.ylabel('Persone', fontsize=16)
plt.title(u"Qual è il tuo titolo di studio?", fontsize=18, y=1.02)
my_colors = seaborn.color_palette("husl", len(study)) # Set color palette
study.plot(kind="bar",color=my_colors)
plt.savefig("svg/Q007-MakerTitoliDiStudio01.svg")
plt.savefig("png/Q007-MakerTitoliDiStudio01.png")
plt.savefig("pdf/Q007-MakerTitoliDiStudio01.pdf")
```
