# Q039 - Qual é il tuo approccio alla progettazione?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html

# Change the font
matplotlib.rcParams.update({'font.family': 'Source Sans Pro'})
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q039[SQ001] - Q039[SQ005] - Q039[other]

approach_columns = ['Q039[SQ001]','Q039[SQ002]','Q039[SQ003]','Q039[SQ004]','Q039[SQ005]']
approach_options = ['Creo da zero i miei progetti',
                  'Utilizzo e modifico progetti esistenti (open design)',
                  'Sviluppo progetti in collaborazione con una community (codesign)',
                  'Utilizzo applicazioni che generano modelli (design generativo)',
                  'Sviluppo il progetto mentre lo realizzo materialmente (tinkering)']
approach = data[approach_columns]
approach.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
approach_other = data['Q039[other]'].str.lower().value_counts()
```

```python
#approach[0:4]
```

```python
# Gather data
approach_b = {}
for k,i in enumerate(approach_columns):
    approach_b[k] = approach[i].value_counts(dropna=False)
    print "Data:",approach_options[k]
    print approach_b[k]
    print
    print "Data %:",approach_options[k]
    print approach[i].value_counts(normalize=True,dropna=False)*100
    print
    
yes = []
no = []
nanvalue = []

for k,i in enumerate(approach_columns):
    approach_presents = approach_b[k].index.tolist()
    
    # Convert NaN to "NaN"
    for o,h in enumerate(approach_presents):
        if type(h) is float:
            approach_presents.pop(o)
            approach_presents.append("NaN")
    
    # Reassign new list with "NaN"
    approach_b[k].index = approach_presents
    
    # Check for empty values, and put a 0 instead
    if "Si" not in approach_presents:
        yes.append(0)
    if "No" not in approach_presents:
        no.append(0)
    if "NaN" not in approach_presents:
        nanvalue.append(0)
    
    for j in approach_presents:
        if j == "Si":
            yes.append(approach_b[k].ix["Si"])
        elif j == "No":
            no.append(approach_b[k].ix["No"])
        elif j == "NaN":
            nanvalue.append(approach_b[k].ix["NaN"]) 
```

```python
# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel('Approcci', fontsize=16)
plt.ylabel('Persone', fontsize=16)
plt.title(u'Qual é il tuo approccio alla progettazione?', fontsize=18)
plt.xticks(range(len(approach_options)+1),approach_options,rotation=90)
ind = np.arange(len(approach_columns))   # the x locations for the groups
width = 0.25                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 3) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[1],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[0],align='center') # Plot No 
rect3 = plt.bar(ind+width*2,nanvalue,width,color=my_colors[2],align='center') # Plot NaN
plt.legend( (rect1, rect2, rect3), ('Si', 'No', 'NaN') )
plt.savefig("svg/Q039-MakerApprocciAllaProgettazione01.svg")
plt.savefig("png/Q039-MakerApprocciAllaProgettazione01.png")
```
