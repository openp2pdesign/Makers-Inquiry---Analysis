# Q039 - Qual é il tuo approccio alla progettazione?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q039[SQ001] - Q039[SQ005] - Q039[other]

approach_columns = ['Q039[SQ001]','Q039[SQ002]','Q039[SQ003]','Q039[SQ004]','Q039[SQ005]']
approach_options = ['Creo da zero i miei progetti',
                  'Utilizzo e modifico progetti esistenti (open design)',
                  'Sviluppo progetti in collaborazione con una community (codesign)',
                  'Utilizzo applicazioni che generano modelli (design generativo)',
                  'Sviluppo il progetto mentre lo realizzo materialmente (tinkering)']
approach = data[approach_columns]
approach.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
approach_other = data['Q039[other]'].str.lower().value_counts()
```

```python
approach[0:4]
```

```python
# Gather data
approach_b = {}
for k,i in enumerate(approach_columns):
    approach_b[k] = approach[i].value_counts()
    print "Data:",approach_options[k]
    print approach_b[k]
    print
    print "Data %:",approach_options[k]
    print approach[i].value_counts(normalize=True)*100
    print
    
yes = []
no = []

# Get data for plotting, and check that we have all the values for Yes and No
for k,i in enumerate(approach_columns):
    if len(approach_b[k].index) == 2:
        if approach_b[k].index[0] == "Si":
            yes.append(approach_b[k].iloc[0])
            no.append(approach_b[k].iloc[1])
        elif approach_b[k].index[1] == "Si":
            yes.append(approach_b[k].iloc[1])
            no.append(approach_b[k].iloc[0])
    elif len(approach_b[k].index) == 1:
        if approach_b[k].index[0] == "Si":
            yes.append(approach_b[k].iloc[0])
            no.append(0)
        elif approach_b[k].index[0] == "No":
            yes.append(0)
            no.append(approach_b[k].iloc[0])
    else:
        yes.append(0)
        no.append(0)

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel('Approcci')
plt.ylabel('Persone')
plt.title('Maker: Approcci alla progettazione')
plt.xticks(range(len(approach_options)),approach_options,rotation=90)
ind = np.arange(len(approach_columns))   # the x locations for the groups
width = 0.35                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 2) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[1],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[0],align='center') # Plot No 
plt.legend( (rect1, rect2), ('Si', 'No') )
plt.savefig("pdf/Q039-MakerApprocciAllaProgettazione01.pdf")
```
