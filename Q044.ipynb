# Q044 - Che tipo di progetti o prodotti realizzi?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html

# Change the font
matplotlib.rcParams.update({'font.family': 'Source Sans Pro'})
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q044[SQ001] - Q044[SQ009]

project_type_columns = ['Q044[SQ001]','Q044[SQ002]','Q044[SQ003]','Q044[SQ004]','Q044[SQ005]','Q044[SQ006]','Q044[SQ007]','Q044[SQ008]','Q044[SQ009]']
project_type_options = ['Prototipi',
                  'Prodotti unici',
                  'Prodotti personalizzati',
                  'Prodotti hackerati',
                  'Prodotti rigenerati',
                  'Prodotti riparati',
                  'Prodotti in edizioni limitate',
                  'Prodotto in piccola serie',
                  'Programmi e applicazioni (software)']
project_type = data[project_type_columns]
project_type.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
```

```python
#project_type[0:4]
```

```python
# Gather data
project_type_b = {}
for k,i in enumerate(project_type_columns):
    project_type_b[k] = project_type[i].value_counts(dropna=False)
    print "Data:",project_type_options[k]
    print project_type_b[k]
    print
    print "Data %:",project_type_options[k]
    print project_type[i].value_counts(normalize=True,dropna=False)*100
    print
    
yes = []
no = []
nanvalue = []

for k,i in enumerate(project_type_columns):
    project_type_presents = project_type_b[k].index.tolist()
    
    # Convert NaN to "NaN"
    for o,h in enumerate(project_type_presents):
        if type(h) is float:
            project_type_presents.pop(o)
            project_type_presents.append("NaN")
    
    # Reassign new list with "NaN"
    project_type_b[k].index = project_type_presents
    
    # Check for empty values, and put a 0 instead
    if "Si" not in project_type_presents:
        yes.append(0)
    if "No" not in project_type_presents:
        no.append(0)
    if "NaN" not in project_type_presents:
        nanvalue.append(0)
    
    for j in project_type_presents:
        if j == "Si":
            yes.append(project_type_b[k].ix["Si"])
        elif j == "No":
            no.append(project_type_b[k].ix["No"])
        elif j == "NaN":
            nanvalue.append(project_type_b[k].ix["NaN"]) 
```

```python
# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel('Tipi di progetti', fontsize=16)
plt.ylabel('Persone', fontsize=16)
plt.title('Che tipo di progetti o prodotti realizzi?', fontsize=18)
plt.xticks(range(len(project_type_options)+1),project_type_options,rotation=90)
ind = np.arange(len(project_type_columns))   # the x locations for the groups
width = 0.25                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 3) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[1],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[0],align='center') # Plot No 
rect3 = plt.bar(ind+width*2,nanvalue,width,color=my_colors[2],align='center') # Plot NaN
plt.legend( (rect1, rect2, rect3), ('Si', 'No', 'NaN') )
plt.savefig("svg/Q044-MakerTipiDiProgettiProdotti01.svg")
plt.savefig("png/Q044-MakerTipiDiProgettiProdotti01.png")
```
