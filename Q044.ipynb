# Q044 - Che tipo di progetti o prodotti realizzi?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q044[SQ001] - Q044[SQ009]

project_type_columns = ['Q044[SQ001]','Q044[SQ002]','Q044[SQ003]','Q044[SQ004]','Q044[SQ005]','Q044[SQ006]','Q044[SQ007]','Q044[SQ008]','Q044[SQ009]']
project_type_options = ['Prototipi',
                  'Prodotti unici',
                  'Prodotti personalizzati',
                  'Prodotti hackerati',
                  'Prodotti rigenerati',
                  'Prodotti riparati',
                  'Prodotti in edizioni limitate',
                  'Prodotto in piccola serie',
                  'Programmi e applicazioni (software)']
project_type = data[project_type_columns]
project_type.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
```

```python
project_type[0:4]
```

```python
# Gather data
project_type_b = {}
for k,i in enumerate(project_type_columns):
    project_type_b[k] = project_type[i].value_counts()
    print "Data:",project_type_options[k]
    print project_type_b[k]
    print
    print "Data %:",project_type_options[k]
    print project_type[i].value_counts(normalize=True)*100
    print
    
yes = []
no = []

# Get data for plotting, and check that we have all the values for Yes and No
for k,i in enumerate(project_type_columns):
    if len(project_type_b[k].index) == 2:
        if project_type_b[k].index[0] == "Si":
            yes.append(project_type_b[k].iloc[0])
            no.append(project_type_b[k].iloc[1])
        elif project_type_b[k].index[1] == "Si":
            yes.append(project_type_b[k].iloc[1])
            no.append(project_type_b[k].iloc[0])
    elif len(project_type_b[k].index) == 1:
        if project_type_b[k].index[0] == "Si":
            yes.append(project_type_b[k].iloc[0])
            no.append(0)
        elif project_type_b[k].index[0] == "No":
            yes.append(0)
            no.append(project_type_b[k].iloc[0])
    else:
        yes.append(0)
        no.append(0)

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel('Tipi di progetti')
plt.ylabel('Persone')
plt.title('Maker: Tipi di progetti e prodotti realizzati')
plt.xticks(range(len(project_type_options)),project_type_options,rotation=90)
ind = np.arange(len(project_type_columns))   # the x locations for the groups
width = 0.35                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 2) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[1],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[0],align='center') # Plot No 
plt.legend( (rect1, rect2), ('Si', 'No') )
plt.savefig("pdf/Q044-MakerTipiDiProgettiProdotti01.pdf")
```
