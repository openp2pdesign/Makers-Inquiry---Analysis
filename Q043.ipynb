# Q043 - Come vendi i tuoi prodotti?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html

# Change the font
matplotlib.rcParams.update({'font.family': 'Source Sans Pro'})
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q043[SQ001] - Q043[SQ006] - Q039[other]

sale_columns = ['Q043[SQ001]','Q043[SQ002]','Q043[SQ003]','Q043[SQ004]','Q043[SQ005]','Q043[SQ006]']
sale_options = ['Fiere',
                  'Mostre-mercato',
                  'Personal marketplace',
                  'Piattaforme e-commerce',
                  'Passaparola',
                  'Negozio (anche temporary shop)']
sale = data[sale_columns]
sale.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
sale_other = data['Q043[other]'].str.lower().value_counts()
```

```python
#sale[0:4]
```

```python
# Gather data
sale_b = {}
for k,i in enumerate(sale_columns):
    sale_b[k] = sale[i].value_counts(dropna=False)
    print "Data:",sale_options[k]
    print sale_b[k]
    print
    print "Data %:",sale_options[k]
    print sale[i].value_counts(normalize=True,dropna=False)*100
    print
    
yes = []
no = []
nanvalue = []

for k,i in enumerate(sale_columns):
    sale_presents = sale_b[k].index.tolist()
    
    # Convert NaN to "NaN"
    for o,h in enumerate(sale_presents):
        if type(h) is float:
            sale_presents.pop(o)
            sale_presents.append("NaN")
    
    # Reassign new list with "NaN"
    sale_b[k].index = sale_presents
    
    # Check for empty values, and put a 0 instead
    if "Si" not in sale_presents:
        yes.append(0)
    if "No" not in sale_presents:
        no.append(0)
    if "NaN" not in sale_presents:
        nanvalue.append(0)
    
    for j in sale_presents:
        if j == "Si":
            yes.append(sale_b[k].ix["Si"])
        elif j == "No":
            no.append(sale_b[k].ix["No"])
        elif j == "NaN":
            nanvalue.append(sale_b[k].ix["NaN"]) 
```

```python
# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel('Modi di vendita', fontsize=16)
plt.ylabel('Persone', fontsize=16)
plt.title('Come vendi i tuoi prodotti?', fontsize=18, y=1.02)
plt.xticks(range(len(sale_options)+1),sale_options,rotation=90)
ind = np.arange(len(sale_columns))   # the x locations for the groups
width = 0.25                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 3) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[1],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[0],align='center') # Plot No 
rect3 = plt.bar(ind+width*2,nanvalue,width,color=my_colors[2],align='center') # Plot NaN
plt.legend( (rect1, rect2, rect3), ('Si', 'No', 'NaN') )
plt.savefig("svg/Q043-MakerComeVendi01.svg")
plt.savefig("png/Q043-MakerComeVendi01.png")
plt.savefig("pdf/Q043-MakerComeVendi01.pdf")
```
