```python
# Range: Q027[SQ001] - Q026[SQ011] - Q027[other]

meaning_columns = ['Q027[SQ001]','Q027[SQ002]','Q027[SQ003]','Q027[SQ004]','Q027[SQ005]','Q027[SQ006]',
                       'Q027[SQ007]','Q027[SQ008]','Q027[SQ009]','Q027[SQ010]','Q027[SQ011]']
meaning_options = [u"DIY",u"Autoproduzione",u"Digital Fabrication (Fabbing)",u"Fab Lab e makerspace",
                   u"Physical computing e tinkering",u"Openness (Open Design, Open Hardware,...)",
                   u"Collaboration e sharing",u"Passatempo",u"Nel tempo libero",u"Come attivitá ricreativa",
                   u"Come attivitá di consumo"]

meaning = data[meaning_columns]
meaning.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
meaning_other = data['Q027[other]'].str.lower().value_counts()
```

```python
meaning[0:4]
```

```python
# Gather data
meaning_b = {}
for k,i in enumerate(meaning_columns):
    meaning_b[k] = meaning[i].value_counts()
    print "Data:",meaning_options[k]
    print meaning_b[k]
    print
    print "Data %:",meaning_options[k]
    print meaning[i].value_counts(normalize=True)*100
    print
    
yes = []
no = []

# Get data for plotting, and check that we have all the values for Yes and No
for k,i in enumerate(meaning_columns):
    if len(meaning_b[k].index) == 2:
        if meaning_b[k].index[0] == "Si":
            yes.append(meaning_b[k].iloc[0])
            no.append(meaning_b[k].iloc[1])
        elif meaning_b[k].index[1] == "Si":
            yes.append(meaning_b[k].iloc[1])
            no.append(meaning_b[k].iloc[0])
    elif len(meaning_b[k].index) == 1:
        if meaning_b[k].index[0] == "Si":
            yes.append(meaning_b[k].iloc[0])
            no.append(0)
        elif meaning_b[k].index[0] == "No":
            yes.append(0)
            no.append(meaning_b[k].iloc[0])
    else:
        yes.append(0)
        no.append(0)

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel('Significati')
plt.ylabel('Persone')
plt.title('Maker: Significati del Making')
plt.xticks(range(len(meaning_options)),meaning_options,rotation=90)
ind = np.arange(len(meaning_columns))   # the x locations for the groups
width = 0.35                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 2) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[0],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[1],align='center') # Plot No 
plt.legend( (rect1, rect2), ('Si', 'No') )
plt.savefig("pdf/Q027-MakerSignificatiMaking01.pdf")
```
