# Q027 - Quali sono le motivazioni che ti spingono a occuparti di making?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html

# Change the font
matplotlib.rcParams.update({'font.family': 'Source Sans Pro'})
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q027[SQ001] - Q027[SQ008] - Q027[other]

motivations_columns = ['Q027[SQ001]','Q027[SQ002]','Q027[SQ003]','Q027[SQ004]','Q027[SQ005]','Q027[SQ006]','Q027[SQ007]','Q027[SQ008]']
motivations_options = [u"Voglio sperimentare",u"Voglio apprendere",u"Voglio divertirmi",u"Voglio creare un nuovo prodotto/servizio o un’impresa",
u"Voglio conoscere persone",u"Voglio collaborare con altri", 
u"Voglio dedicarmi a progetti che hanno un impatto positivo sulla comunità locale",
u"Voglio trovare soluzioni alternative ai modelli di consumo vigenti"]

motivations = data[motivations_columns]
motivations.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
motivations_other = data['Q027[other]'].str.lower().value_counts()
```

```python
#motivations[0:4]
```

```python
# Gather data
motivations_b = {}
for k,i in enumerate(motivations_columns):
    motivations_b[k] = motivations[i].value_counts(dropna=False)
    print "Data:",motivations_options[k]
    print motivations_b[k]
    print
    print "Data %:",motivations_options[k]
    print motivations[i].value_counts(normalize=True,dropna=False)*100
    print
    
yes = []
no = []
nanvalue = []

for k,i in enumerate(motivations_columns):
    motivations_presents = motivations_b[k].index.tolist()
    
    # Convert NaN to "NaN"
    for o,h in enumerate(motivations_presents):
        if type(h) is float:
            motivations_presents.pop(o)
            motivations_presents.append("NaN")
    
    # Reassign new list with "NaN"
    motivations_b[k].index = motivations_presents
    
    # Check for empty values, and put a 0 instead
    if "Si" not in motivations_presents:
        yes.append(0)
    if "No" not in motivations_presents:
        no.append(0)
    if "NaN" not in motivations_presents:
        nanvalue.append(0)
    
    for j in motivations_presents:
        if j == "Si":
            yes.append(motivations_b[k].ix["Si"])
        elif j == "No":
            no.append(motivations_b[k].ix["No"])
        elif j == "NaN":
            nanvalue.append(motivations_b[k].ix["NaN"]) 
```

```python
# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel('Motivi', fontsize=16)
plt.ylabel('Persone', fontsize=16)
plt.title('Quali sono le motivazioni che ti spingono a occuparti di making?', fontsize=18)
plt.xticks(range(len(motivations_options)+1),motivations_options,rotation=90)
ind = np.arange(len(motivations_columns))   # the x locations for the groups
width = 0.35                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 3) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[0],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[1],align='center') # Plot No
rect3 = plt.bar(ind+width*2,nanvalue,width,color=my_colors[2],align='center') # Plot NaN
plt.legend( (rect1, rect2, rect3), ('Si', 'No', 'NaN') )
plt.savefig("svg/Q027-MakerMotiviMaking01.svg")
plt.savefig("png/Q027-MakerMotiviMaking01.png")
```
