# Q027 - Quali sono le motivazioni che ti spingono a occuparti di making?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q027[SQ001] - Q027[SQ008] - Q027[other]

motivations_columns = ['Q027[SQ001]','Q027[SQ002]','Q027[SQ003]','Q027[SQ004]','Q027[SQ005]','Q027[SQ006]','Q027[SQ007]','Q027[SQ008]']
motivations_options = [u"Voglio sperimentare",u"Voglio apprendere",u"Voglio divertirmi",u"Voglio creare un nuovo prodotto/servizio o un’impresa",
u"Voglio conoscere persone",u"Voglio collaborare con altri", 
u"Voglio dedicarmi a progetti che hanno un impatto positivo sulla comunità locale",
u"Voglio trovare soluzioni alternative ai modelli di consumo vigenti"]

motivations = data[motivations_columns]
motivations.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
motivations_other = data['Q027[other]'].str.lower().value_counts()
```

```python
motivations[0:4]
```

```python
# Gather data
motivations_b = {}
for k,i in enumerate(motivations_columns):
    motivations_b[k] = motivations[i].value_counts()
    print "Data:",motivations_options[k]
    print motivations_b[k]
    print
    print "Data %:",motivations_options[k]
    print motivations[i].value_counts(normalize=True)*100
    print
    
yes = []
no = []

# Get data for plotting, and check that we have all the values for Yes and No
for k,i in enumerate(motivations_columns):
    if len(motivations_b[k].index) == 2:
        if motivations_b[k].index[0] == "Si":
            yes.append(motivations_b[k].iloc[0])
            no.append(motivations_b[k].iloc[1])
        elif motivations_b[k].index[1] == "Si":
            yes.append(motivations_b[k].iloc[1])
            no.append(motivations_b[k].iloc[0])
    elif len(motivations_b[k].index) == 1:
        if motivations_b[k].index[0] == "Si":
            yes.append(motivations_b[k].iloc[0])
            no.append(0)
        elif motivations_b[k].index[0] == "No":
            yes.append(0)
            no.append(motivations_b[k].iloc[0])
    else:
        yes.append(0)
        no.append(0)

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel('Motivi')
plt.ylabel('Persone')
plt.title('Maker: Motivi per il Making')
plt.xticks(range(len(motivations_options)),motivations_options,rotation=90)
ind = np.arange(len(motivations_columns))   # the x locations for the groups
width = 0.35                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 2) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[0],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[1],align='center') # Plot No 
plt.legend( (rect1, rect2), ('Si', 'No') )
plt.savefig("pdf/Q027-MakerMotiviMaking01.pdf")
```
