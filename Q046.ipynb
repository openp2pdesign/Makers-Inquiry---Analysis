# Q046 - Nel caso di prodotti materiali quanti ne realizzi annualmente?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Get the distribution of way of living
product_output = data["Q046"].value_counts(dropna=False)
print "Data:"
print product_output
print ""
print "Data %:"
print data["Q046"].value_counts(normalize=True,dropna=False) * 100

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Quantitá produzione')
plt.ylabel('Persone')
my_colors = seaborn.color_palette("husl", len(product_output)) # Set color palette
product_output.plot(kind="bar",title="Nel caso di prodotti materiali quanti ne realizzi annualmente?",color=my_colors)
plt.savefig("pdf/Q046-MakerQuantoProduci01.pdf") # Save pdf
```

```python
# Order of the choices
product_output_order = [u"Meno di 5 unità",u"Meno di 10 unità",u"Meno di 50 unità",u"Meno di 100 unità",u"Oltre 100 unità"]

product_output2 = product_output.reindex(product_output_order)

# Get the distribution of way of living, reindexed
print "Data:"
print product_output2
print ""
print "Data %:"
product_output2_normalized = data["Q046"].value_counts(normalize=True) * 100
print product_output2_normalized.reindex(product_output_order)

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Quantitá produzione')
plt.ylabel('Persone')
my_colors = seaborn.color_palette("husl", len(product_output2)) # Set color palette
product_output2.plot(kind="bar",title="Nel caso di prodotti materiali quanti ne realizzi annualmente?",color=my_colors)
plt.savefig("pdf/Q046-MakerQuantoProduci02.pdf") # Save pdf
```
