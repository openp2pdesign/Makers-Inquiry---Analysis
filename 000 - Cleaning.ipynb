```python
# Check for NaN data
check = pd.isnull(data)
check.head()
```

```python
responses = {}

# Check all the complete and incomplete responses
for i,n in enumerate(check.index):
    responses[n] = {}
    responses[n]["complete"] = 0
    responses[n]["incomplete"] = 0

    for j in check.columns:
        if check.ix[i][j] == False:
            responses[n]["complete"] += 1
        else:
            responses[n]["incomplete"] += 1
```

```python
# Transform data to lists for plotting
ids = list(responses)
yes = []
no = []
for i in responses:
    yes.append(responses[i]["complete"])
    no.append(responses[i]["incomplete"])

# Plot the data
plt.figure(figsize=(142,30))
plt.xlabel('Number')
plt.ylabel('Responses')
plt.title('Survey: complete vs incomplete responses')
plt.xticks(range(len(responses)),ids,rotation=0)
ind = np.arange(len(responses))   # the x locations for the groups
width = 0.35                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 2) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[1],align='center') # Plot Yes
#rect2 = plt.bar(ind+width,no,width,color=my_colors[0],align='center') # Plot No 
#plt.legend( (rect1, rect2), ('Complete', 'Incomplete') )
plt.savefig("pdf/000-CompleteVsIncompleteResponses.pdf")
```

## Alternative method: filter by percentage of completeness

```python
# Filter the data according to a percentage of complete responses
surveys = len(responses)
questions = len(data.columns)

# The new clean dataframe
cleaned_data = pd.DataFrame(columns=data.columns)

for i in data.index:
    # Calculate the percentage
    completeness = float(responses[i]["complete"]) / questions * 100
    # Round the percentage
    completeness = float("{0:.2f}".format(completeness))
    
    # Accept or reject
    if completeness > 40:
        print i,"- ACCEPTED! Question n.",i,"is",completeness,"% complete"
        cleaned_data = cleaned_data.append(data.ix[i])
    else:
        print i,"- REJECTED! Question n.",i,"is",completeness,"% complete"
```
