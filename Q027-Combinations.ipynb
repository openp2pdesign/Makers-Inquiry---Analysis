```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html

# Change the font
matplotlib.rcParams.update({'font.family': 'Source Sans Pro'})
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q027[SQ001] - Q027[SQ008] - Q027[other]

motivations_columns = ['Q027[SQ001]','Q027[SQ002]','Q027[SQ003]','Q027[SQ004]','Q027[SQ005]','Q027[SQ006]','Q027[SQ007]','Q027[SQ008]']
motivations_options = [u"Voglio sperimentare",u"Voglio apprendere",u"Voglio divertirmi",u"Voglio creare un nuovo prodotto/servizio o un’impresa",
u"Voglio conoscere persone",u"Voglio collaborare con altri", 
u"Voglio dedicarmi a progetti che hanno un impatto positivo sulla comunità locale",
u"Voglio trovare soluzioni alternative ai modelli di consumo vigenti"]

motivations = data[motivations_columns]
motivations.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
motivations_other = data['Q027[other]'].str.lower().value_counts()
```

```python
#motivations[0:4]
```

## Combinations...

```python
# Create all the possible combinations from the main options
# See http://stackoverflow.com/questions/17176887/python-get-all-permutation-of-a-list-w-o-repetitions

import itertools 

all_combinations = {}
all_combinations_columns = []

for i in range(1, len(motivations_columns)+1):
    comb = list(itertools.combinations(motivations_columns, i))
    for k in comb:
        # Each combination
        all_combinations[k] = {}
        all_combinations[k]["col_list"] = list(k)
        # Build the string and boolean list of each combination
        comb_list = []
        comb_bool_list = []
        # Put default False value
        for l in motivations_columns:
            comb_bool_list.append(False)
        for j in k:
            pos = motivations_columns.index(j) # Get position
            comb_list.append(motivations_options[pos])
            comb_bool_list[pos] = True
        all_combinations[k]["list"] = comb_list
        all_combinations[k]["bool_list"] = comb_bool_list
        all_combinations[k]["str"] = ", ".join(comb_list)
```

```python
# Check which combinations correspond each row
str_values = []
for i in motivations.index:
    current_bool_list = list(motivations.ix[i].isin(["Si"]))
    for i in all_combinations:
        if current_bool_list == all_combinations[i]["bool_list"]:
            str_values.append(all_combinations[i]["str"])
```

```python
# Add combination column
motivations["Combination"] = pd.Series(str_values)
```

```python
# Gather data
resulting_combinations = motivations["Combination"].value_counts()
resulting_combinations_percentage = motivations["Combination"].value_counts(normalize=True)*100
```

```python
# Plotting the first 10 values of the most popular combinations
resulting_combinations[0:10].plot(kind='bar',figsize=(20,10),rot=90)
plt.title("Quali sono le motivazioni che ti spingono a occuparti di making?", fontsize=18)
plt.ylabel("Persone", fontsize=16)
plt.xlabel("Combinazioni", fontsize=16)
plt.savefig("svg/Q027-Combinazioni.svg")
plt.savefig("png/Q027-Combinazioni.png")
```
