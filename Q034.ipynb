# Q034 - Quante persone lo frequentano?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html

# Change the font
matplotlib.rcParams.update({'font.family': 'Source Sans Pro'})
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Get the distribution of contract type
contract = data["Q034"].value_counts(dropna=False)
print "Data:"
print contract
print ""
print "Data %:"
print data["Q034"].value_counts(normalize=True,dropna=False) * 100

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Frequenza', fontsize=16)
plt.ylabel('Persone', fontsize=16)
plt.title("Quante persone lo frequentano?", fontsize=18)
my_colors = seaborn.color_palette("husl", len(contract)) # Set color palette
contract.plot(kind="bar",color=my_colors)
plt.savefig("svg/Q034-PersoneFrequentanoLab01.svg")
plt.savefig("png/Q034-PersoneFrequentanoLab01.png")
```

```python
# Get the distribution of contract type
contract = data["Q034"].value_counts()
print "Data:"
print contract
print ""
print "Data %:"
print data["Q034"].value_counts(normalize=True) * 100

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Frequenza', fontsize=16)
plt.ylabel('Persone', fontsize=16)
plt.title("Quante persone lo frequentano?", fontsize=18)
my_colors = seaborn.color_palette("husl", len(contract)) # Set color palette
contract.plot(kind="bar",color=my_colors)
plt.savefig("svg/Q034-PersoneFrequentanoLab02.svg")
plt.savefig("png/Q034-PersoneFrequentanoLab02.png")
```
