# Q054 - Quali fra i seguenti elementi costituiscono la tua offerta di servizi?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q054[SQ001] - Q054[SQ010] - Q059[other]

services_columns = ['Q054[SQ001]','Q054[SQ002]','Q054[SQ003]','Q054[SQ004]','Q054[SQ005]',
                    'Q054[SQ006]','Q054[SQ007]','Q054[SQ008]','Q054[SQ009]','Q054[SQ010]']
services_options = ['Servizi per la progettazione on demand (workshop)',
                    'Servizi per la fabbricazione on demand',
                    'Servizi per la formazione (corsi e workshop didattici)',
                    'Servizi per la ricerca',
                    'Servizi per la progettazione di sistemi-prodotto (workshop progettuali)',
                    'Servizi per la progettazione di software (coding)',
                    'Servizi per la progettazione di tecnologie e macchine utensili',
                    'Servizi per la riparazione, rigenerazione e manutenzione di prodotti e tecnologie',
                    'Organizzazione di eventi culturali',
                    'Spazi Coworking'                    
                    ]
services = data[services_columns]
services.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
services_other = data['Q054[other]'].str.lower().value_counts()
```

```python
#services[0:4]
```

```python
# Gather data
services_b = {}

# Get rid of the people who did not answer to this section
services = services.dropna()

for k,i in enumerate(services_columns):
    services_b[k] = services[i].value_counts(dropna=False)
    print "Data:",services_options[k]
    print services_b[k]
    print
    print "Data %:",services_options[k]
    print services[i].value_counts(normalize=True,dropna=False)*100
    print
    
yes = []
no = []
nanvalue = []

for k,i in enumerate(services_columns):
    services_presents = services_b[k].index.tolist()
    
    # Convert NaN to "NaN"
    for o,h in enumerate(services_presents):
        if type(h) is float:
            services_presents.pop(o)
            services_presents.append("NaN")
    
    # Reassign new list with "NaN"
    services_b[k].index = services_presents
    
    # Check for empty values, and put a 0 instead
    if "Si" not in services_presents:
        yes.append(0)
    if "No" not in services_presents:
        no.append(0)
    if "NaN" not in services_presents:
        nanvalue.append(0)
    
    for j in services_presents:
        if j == "Si":
            yes.append(services_b[k].ix["Si"])
        elif j == "No":
            no.append(services_b[k].ix["No"])
        elif j == "NaN":
            nanvalue.append(services_b[k].ix["NaN"]) 
```

```python
# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Servizi')
plt.ylabel('Persone')
plt.title(u'Quali fra i seguenti elementi costituiscono la tua offerta di servizi?')
plt.xticks(range(len(services_options)+1),services_options,rotation=90)
ind = np.arange(len(services_columns))   # the x locations for the groups
width = 0.25                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 3) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[1],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[0],align='center') # Plot No 
plt.legend( (rect1, rect2), ('Si', 'No') )
plt.savefig("pdf/Q054-MakerspaceServices01.pdf")
```
