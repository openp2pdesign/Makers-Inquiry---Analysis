# Q042 - Se non possiedi le tecnologie che utilizzi per la tua attivitá di making, come fai ad usarle?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q042[SQ001] - Q042[SQ004] - Q042[other]

access_columns = ['Q042[SQ001]','Q042[SQ002]','Q042[SQ003]','Q042[SQ004]']
access_options = ['Prestito o affitto',
                  'Acquisto di servizi (es. Ponoko)',
                  'Accesso a strutture produttive (es. FabLab)',
                  'Accesso a network (es. 3D Hubs)']
access = data[access_columns]
access.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
access_other = data['Q042[other]'].str.lower().value_counts()
```

```python
access[0:4]
```

```python
# Gather data
access_b = {}
for k,i in enumerate(access_columns):
    access_b[k] = access[i].value_counts()
    print "Data:",access_options[k]
    print access_b[k]
    print
    print "Data %:",access_options[k]
    print access[i].value_counts(normalize=True)*100
    print
    
yes = []
no = []

# Get data for plotting, and check that we have all the values for Yes and No
for k,i in enumerate(access_columns):
    if len(access_b[k].index) == 2:
        if access_b[k].index[0] == "Si":
            yes.append(access_b[k].iloc[0])
            no.append(access_b[k].iloc[1])
        elif access_b[k].index[1] == "Si":
            yes.append(access_b[k].iloc[1])
            no.append(access_b[k].iloc[0])
    elif len(access_b[k].index) == 1:
        if access_b[k].index[0] == "Si":
            yes.append(access_b[k].iloc[0])
            no.append(0)
        elif access_b[k].index[0] == "No":
            yes.append(0)
            no.append(access_b[k].iloc[0])
    else:
        yes.append(0)
        no.append(0)

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Modalitá di accesso')
plt.ylabel('Persone')
plt.title('Accesso a tecnologie')
plt.xticks(range(len(access_options)),access_options,rotation=90)
ind = np.arange(len(access_columns))   # the x locations for the groups
width = 0.35                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 2) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[1],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[0],align='center') # Plot No 
plt.legend( (rect1, rect2), ('Si', 'No') )
plt.savefig("pdf/Q042-MakerComeAccedi01.pdf")
```
