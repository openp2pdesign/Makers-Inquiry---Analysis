# Q062 - Cosa chiederesti alle istituzioni per facilitare la tua attivitá di making?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q062[SQ001] - Q062[SQ005] - Q062[other]

policy_columns = ['Q062[SQ001]','Q062[SQ002]','Q062[SQ003]','Q062[SQ004]','Q062[SQ005]']
policy_options = ['Agevolazioni fiscali',
                  'Accesso facilitato al credito',
                  'Semplificazione burocratica',
                  u"Riconoscimento giuridico dell'attivitá di maker",
                  u'Opportunità di finanziamento']
policy = data[policy_columns]
policy.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
policy_other = data['Q062[other]'].str.lower().value_counts()
```

```python
policy[0:4]
```

```python
# Gather data
policy_b = {}
for k,i in enumerate(policy_columns):
    policy_b[k] = policy[i].value_counts(dropna=False)
    print "Data:",policy_options[k]
    print policy_b[k]
    print
    print "Data %:",policy_options[k]
    print policy[i].value_counts(normalize=True,dropna=False)*100
    print
    
yes = []
no = []
nanvalue = []

for k,i in enumerate(policy_columns):
    policy_presents = policy_b[k].index.tolist()
    
    # Convert NaN to "NaN"
    for o,h in enumerate(policy_presents):
        if type(h) is float:
            policy_presents.pop(o)
            policy_presents.append("NaN")
    
    # Reassign new list with "NaN"
    policy_b[k].index = policy_presents
    
    # Check for empty values, and put a 0 instead
    if "Si" not in policy_presents:
        yes.append(0)
    if "No" not in policy_presents:
        no.append(0)
    if "NaN" not in policy_presents:
        nanvalue.append(0)
    
    for j in policy_presents:
        if j == "Si":
            yes.append(policy_b[k].ix["Si"])
        elif j == "No":
            no.append(policy_b[k].ix["No"])
        elif j == "NaN":
            nanvalue.append(policy_b[k].ix["NaN"]) 
```

```python
# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Richieste')
plt.ylabel('Persone')
plt.title(u'Cosa chiederesti alle istituzioni per facilitare la tua attivitá di making?')
plt.xticks(range(len(policy_options)+1),policy_options,rotation=90)
ind = np.arange(len(policy_columns))   # the x locations for the groups
width = 0.25                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 3) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[1],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[0],align='center') # Plot No 
rect3 = plt.bar(ind+width*2,nanvalue,width,color=my_colors[2],align='center') # Plot NaN
rect3 = plt.bar(ind+width*2,nanvalue,width,color=my_colors[2],align='center') # Plot NaN
plt.legend( (rect1, rect2, rect3), ('Si', 'No', 'NaN') )
plt.savefig("pdf/Q062-RichiestePolitiche01.pdf")
```
