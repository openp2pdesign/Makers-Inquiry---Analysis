```python
# Range: Q056[SQ001] - Q056[SQ005] - Q056[other]

offer_columns = ['Q056[SQ001]','Q056[SQ002]','Q056[SQ003]','Q056[SQ004]','Q056[SQ005]']
offer_options = ['Sito web del makerspace','Altre piattaforme web','Social media','Partecipazione a fiere ed eventi','Conoscenze personai (passaparola)']
offer = data[offer_columns]
offer.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
offer_other = data['Q057[other]'].str.lower().value_counts()
```

## Combinations...

```python
# Create all the possible combinations from the main options
# See http://stackoverflow.com/questions/17176887/python-get-all-permutation-of-a-list-w-o-repetitions

import itertools 

all_combinations = {}
all_combinations_columns = []

for i in range(1, len(offer_columns)+1):
    comb = list(itertools.combinations(offer_columns, i))
    for k in comb:
        # Each combination
        all_combinations[k] = {}
        all_combinations[k]["col_list"] = list(k)
        # Build the string and boolean list of each combination
        comb_list = []
        comb_bool_list = []
        # Put default False value
        for l in offer_columns:
            comb_bool_list.append(False)
        for j in k:
            pos = offer_columns.index(j) # Get position
            comb_list.append(offer_options[pos])
            comb_bool_list[pos] = True
        all_combinations[k]["list"] = comb_list
        all_combinations[k]["bool_list"] = comb_bool_list
        all_combinations[k]["str"] = ", ".join(comb_list)
```

```python
# Check which combinations correspond each row
str_values = []
for i in offer.index:
    current_bool_list = list(offer.ix[i].isin(["Si"]))
    for i in all_combinations:
        if current_bool_list == all_combinations[i]["bool_list"]:
            str_values.append(all_combinations[i]["str"])
```

```python
# Add combination column
offer["Combination"] = pd.Series(str_values)
```

```python
# Gather data
resulting_combinations = offer["Combination"].value_counts()
resulting_combinations_percentage = offer["Combination"].value_counts(normalize=True)*100
```

```python
# Plotting the first 10 values of the most popular combinations
resulting_combinations[0:10].plot(kind='bar',figsize=(20,10),rot=90)
plt.savefig("pdf/Q056-Combinazioni.pdf")
```
