```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html

# Change the font
matplotlib.rcParams.update({'font.family': 'Source Sans Pro'})
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q028[SQ001] - Q028[SQ011] - Q028[other]

meaning_columns = ['Q028[SQ001]','Q028[SQ002]','Q028[SQ003]','Q028[SQ004]','Q028[SQ005]','Q028[SQ006]',
                       'Q028[SQ007]','Q028[SQ008]','Q028[SQ009]','Q028[SQ010]','Q028[SQ011]']
meaning_options = [u"DIY",u"Autoproduzione",u"Digital Fabrication (Fabbing)",u"Fab Lab e makerspace",
                   u"Physical computing e tinkering",u"Openness (Open Design, Open Hardware,...)",
                   u"Collaboration e sharing",u"Passatempo",u"Nel tempo libero",u"Come attivitá ricreativa",
                   u"Come attivitá di consumo"]

meaning = data[meaning_columns]
meaning.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
meaning_other = data['Q028[other]'].str.lower().value_counts()
```

## Combinations...

```python
# Create all the possible combinations from the main options
# See http://stackoverflow.com/questions/17176887/python-get-all-permutation-of-a-list-w-o-repetitions

import itertools 

all_combinations = {}
all_combinations_columns = []

for i in range(1, len(meaning_columns)+1):
    comb = list(itertools.combinations(meaning_columns, i))
    for k in comb:
        # Each combination
        all_combinations[k] = {}
        all_combinations[k]["col_list"] = list(k)
        # Build the string and boolean list of each combination
        comb_list = []
        comb_bool_list = []
        # Put default False value
        for l in meaning_columns:
            comb_bool_list.append(False)
        for j in k:
            pos = meaning_columns.index(j) # Get position
            comb_list.append(meaning_options[pos])
            comb_bool_list[pos] = True
        all_combinations[k]["list"] = comb_list
        all_combinations[k]["bool_list"] = comb_bool_list
        all_combinations[k]["str"] = ", ".join(comb_list)
```

```python
# Check which combinations correspond each row
str_values = []
for i in meaning.index:
    current_bool_list = list(meaning.ix[i].isin(["Si"]))
    for i in all_combinations:
        if current_bool_list == all_combinations[i]["bool_list"]:
            str_values.append(all_combinations[i]["str"])
```

```python
# Add combination column
meaning["Combination"] = pd.Series(str_values)
```

```python
# Gather data
resulting_combinations = meaning["Combination"].value_counts()
resulting_combinations_percentage = meaning["Combination"].value_counts(normalize=True)*100
```

```python
# Plotting the first 10 values of the most popular combinations
resulting_combinations[0:10].plot(kind='bar',figsize=(20,10),rot=90)
plt.title("Quali tra le seguenti parole chiave associ al making? Combinazioni", fontsize=16)
plt.ylabel("Persone", fontsize=16)
plt.xlabel("Combinazioni", fontsize=16)
plt.savefig("svg/Q028-Combinazioni.svg")
plt.savefig("png/Q028-Combinazioni.png")
plt.savefig("pdf/Q028-Combinazioni.pdf")
```
