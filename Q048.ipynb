# Q048 - Il risultato della tua attivitá é un'innovazione? Se sí, di che tipo?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q048[SQ001] - Q048[SQ005] - Q048[other]

innovation_columns = ['Q048[SQ001]','Q048[SQ002]','Q048[SQ003]','Q048[SQ004]','Q048[SQ005]']
innovation_options = ['Innovazione di prodotto',
                  'Innovazione di processo (progettazione, produzione, distribuzione)',
                  'Innovazione di sistema-prodotto (prodotto, servizio e processo)',
                  'Innovazione tecnologica',
                  'Innovazione sociale']
innovation = data[innovation_columns]
innovation.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
innovation_other = data['Q048[other]'].str.lower().value_counts()
```

```python
innovation[0:4]
```

```python
# Gather data
innovation_b = {}
for k,i in enumerate(innovation_columns):
    innovation_b[k] = innovation[i].value_counts()
    print "Data:",innovation_options[k]
    print innovation_b[k]
    print
    print "Data %:",innovation_options[k]
    print innovation[i].value_counts(normalize=True)*100
    print
    
yes = []
no = []

# Get data for plotting, and check that we have all the values for Yes and No
for k,i in enumerate(innovation_columns):
    if len(innovation_b[k].index) == 2:
        if innovation_b[k].index[0] == "Si":
            yes.append(innovation_b[k].iloc[0])
            no.append(innovation_b[k].iloc[1])
        elif innovation_b[k].index[1] == "Si":
            yes.append(innovation_b[k].iloc[1])
            no.append(innovation_b[k].iloc[0])
    elif len(innovation_b[k].index) == 1:
        if innovation_b[k].index[0] == "Si":
            yes.append(innovation_b[k].iloc[0])
            no.append(0)
        elif innovation_b[k].index[0] == "No":
            yes.append(0)
            no.append(innovation_b[k].iloc[0])
    else:
        yes.append(0)
        no.append(0)

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Tipi di innovazione')
plt.ylabel('Persone')
plt.title('Che innovazione hanno i tuoi progetti?')
plt.xticks(range(len(innovation_options)),innovation_options,rotation=90)
ind = np.arange(len(innovation_columns))   # the x locations for the groups
width = 0.35                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 2) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[1],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[0],align='center') # Plot No 
plt.legend( (rect1, rect2), ('Si', 'No') )
plt.savefig("pdf/Q048-MakerQualeInnovazione01.pdf")
```
