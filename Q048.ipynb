# Q048 - Il risultato della tua attivitá é un'innovazione? Se sí, di che tipo?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html

# Change the font
matplotlib.rcParams.update({'font.family': 'Source Sans Pro'})
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q048[SQ001] - Q048[SQ005] - Q048[other]

innovation_columns = ['Q048[SQ001]','Q048[SQ002]','Q048[SQ003]','Q048[SQ004]','Q048[SQ005]']
innovation_options = ['Innovazione di prodotto',
                  'Innovazione di processo (progettazione, produzione, distribuzione)',
                  'Innovazione di sistema-prodotto (prodotto, servizio e processo)',
                  'Innovazione tecnologica',
                  'Innovazione sociale']
innovation = data[innovation_columns]
innovation.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
innovation_other = data['Q048[other]'].str.lower().value_counts()
```

```python
#innovation[0:4]
```

```python
# Gather data
innovation_b = {}
for k,i in enumerate(innovation_columns):
    innovation_b[k] = innovation[i].value_counts(dropna=False)
    print "Data:",innovation_options[k]
    print innovation_b[k]
    print
    print "Data %:",innovation_options[k]
    print innovation[i].value_counts(normalize=True,dropna=False)*100
    print
    
yes = []
no = []
nanvalue = []

for k,i in enumerate(innovation_columns):
    innovation_presents = innovation_b[k].index.tolist()
    
    # Convert NaN to "NaN"
    for o,h in enumerate(innovation_presents):
        if type(h) is float:
            innovation_presents.pop(o)
            innovation_presents.append("NaN")
    
    # Reassign new list with "NaN"
    innovation_b[k].index = innovation_presents
    
    # Check for empty values, and put a 0 instead
    if "Si" not in innovation_presents:
        yes.append(0)
    if "No" not in innovation_presents:
        no.append(0)
    if "NaN" not in innovation_presents:
        nanvalue.append(0)
    
    for j in innovation_presents:
        if j == "Si":
            yes.append(innovation_b[k].ix["Si"])
        elif j == "No":
            no.append(innovation_b[k].ix["No"])
        elif j == "NaN":
            nanvalue.append(innovation_b[k].ix["NaN"]) 
```

```python
# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Tipi di innovazione', fontsize=16)
plt.ylabel('Persone', fontsize=16)
plt.title(u"Il risultato della tua attivitá é un'innovazione? Se sí, di che tipo?", fontsize=18)
plt.xticks(range(len(innovation_options)+1),innovation_options,rotation=90)
ind = np.arange(len(innovation_columns))   # the x locations for the groups
width = 0.25                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 3) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[1],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[0],align='center') # Plot No 
rect3 = plt.bar(ind+width*2,nanvalue,width,color=my_colors[2],align='center') # Plot NaN
plt.legend( (rect1, rect2, rect3), ('Si', 'No', 'NaN') )
plt.savefig("svg/Q048-MakerQualeInnovazione01.svg")
plt.savefig("png/Q048-MakerQualeInnovazione01.png")
```
