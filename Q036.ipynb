# Q036 - In quali delle seguenti comunitá online sei coinvolto?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q036[SQ001] - Q036[SQ012] - Q036[other]

communities_columns = ['Q036[SQ001]','Q036[SQ002]','Q036[SQ003]','Q036[SQ004]','Q036[SQ005]',
                    'Q036[SQ006]','Q036[SQ007]','Q036[SQ008]','Q036[SQ009]','Q036[SQ010]','Q036[SQ011]',
                    'Q036[SQ012]','Q036[SQ013]']
communities_options = ['Fabber in Italia (Facebook)',
                    'Associazione Make in Italy (Facebook/sito)',
                    'Fondazione Make in Italy CDB',
                    'Fablabs.io',
                    'Make Magazine',
                    '3D Hubs',
                    'Instructables',
                    'Thingiverse',
                    'International FabLab Association',
                    'wiki.fablab.is',
                    'Open Manufacturing mailing list',
                    'Open Source Hardware Association',
                    'Open Design Working Group @ Open Knowledge Foundation'
                    ]
communities = data[communities_columns]
communities.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
communities_other = data['Q036[other]'].str.lower().value_counts()
```

```python
communities[0:4]
```

```python
# Gather data
communities_b = {}
for k,i in enumerate(communities_columns):
    communities_b[k] = communities[i].value_counts()
    print "Data:",communities_options[k]
    print communities_b[k]
    print
    print "Data %:",communities_options[k]
    print communities[i].value_counts(normalize=True)*100
    print
    
yes = []
no = []

# Get data for plotting, and check that we have all the values for Yes and No
for k,i in enumerate(communities_columns):
    if len(communities_b[k].index) == 2:
        if communities_b[k].index[0] == "Si":
            yes.append(communities_b[k].iloc[0])
            no.append(communities_b[k].iloc[1])
        elif communities_b[k].index[1] == "Si":
            yes.append(communities_b[k].iloc[1])
            no.append(communities_b[k].iloc[0])
    elif len(communities_b[k].index) == 1:
        if communities_b[k].index[0] == "Si":
            yes.append(communities_b[k].iloc[0])
            no.append(0)
        elif communities_b[k].index[0] == "No":
            yes.append(0)
            no.append(communities_b[k].iloc[0])
    else:
        yes.append(0)
        no.append(0)

# Plot the data
plt.figure(figsize=(16,6))
plt.xlabel(u'Comunitá')
plt.ylabel('Persone')
plt.title(u'In quali comunitá online partecipi?')
plt.xticks(range(len(communities_options)),communities_options,rotation=90)
ind = np.arange(len(communities_columns))   # the x locations for the groups
width = 0.35                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 2) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[1],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[0],align='center') # Plot No 
plt.legend( (rect1, rect2), ('Si', 'No') )
plt.savefig("pdf/Q036-communities01.pdf")
```
