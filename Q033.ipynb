# Q033 - Che attivitá svolgi all'interno di questo luogo?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html

# Change the font
matplotlib.rcParams.update({'font.family': 'Source Sans Pro'})
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Range: Q033[SQ001] - Q055[SQ010] - Q055[other]

activities_columns = ['Q033[SQ001]','Q033[SQ002]','Q033[SQ003]','Q033[SQ004]']
activities_options = [u'Seguo alcune delle attivitá organizzate',
                    'Sviluppo i miei progetti',
                    'Sviluppo progetti con altre persone che frequentano lo spazio',
                    'Organizzo iniziative ed eventi con altre persone che frequentano lo spazio'
                    ]
activities = data[activities_columns]
activities.replace(u'Sì', 'Si', inplace=True) # Get rid of accented characters 
activities_other = data['Q055[other]'].str.lower().value_counts()
```

```python
#activities[0:4]
```

```python
# Gather data
activities_b = {}
for k,i in enumerate(activities_columns):
    activities_b[k] = activities[i].value_counts(dropna=False)
    print "Data:",activities_options[k]
    print activities_b[k]
    print
    print "Data %:",activities_options[k]
    print activities[i].value_counts(normalize=True,dropna=False)*100
    print
    
yes = []
no = []
nanvalue = []

for k,i in enumerate(activities_columns):
    activities_presents = activities_b[k].index.tolist()
    
    # Convert NaN to "NaN"
    for o,h in enumerate(activities_presents):
        if type(h) is float:
            activities_presents.pop(o)
            activities_presents.append("NaN")
    
    # Reassign new list with "NaN"
    activities_b[k].index = activities_presents
    
    # Check for empty values, and put a 0 instead
    if "Si" not in activities_presents:
        yes.append(0)
    if "No" not in activities_presents:
        no.append(0)
    if "NaN" not in activities_presents:
        nanvalue.append(0)
    
    for j in activities_presents:
        if j == "Si":
            yes.append(activities_b[k].ix["Si"])
        elif j == "No":
            no.append(activities_b[k].ix["No"])
        elif j == "NaN":
            nanvalue.append(activities_b[k].ix["NaN"]) 
```

```python
# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Attivitá', fontsize=16)
plt.ylabel('Persone', fontsize=16)
plt.title(u"Che attivitá svolgi all'interno di questo luogo?", fontsize=18)
plt.xticks(range(len(activities_options)+1),activities_options,rotation=90)
ind = np.arange(len(activities_columns))   # the x locations for the groups
width = 0.25                              # the width of the bars

my_colors = seaborn.color_palette("Set1", 5) # Set color palette
rect1 = plt.bar(ind,yes,width,color=my_colors[1],align='center') # Plot Yes
rect2 = plt.bar(ind+width,no,width,color=my_colors[0],align='center') # Plot No 
rect3 = plt.bar(ind+width*2,nanvalue,width,color=my_colors[2],align='center') # Plot NaN
plt.legend( (rect1, rect2, rect3), ('Si', 'No', 'NaN') )
plt.savefig("svg/Q033-activities01.svg")
plt.savefig("png/Q033-activities01.png")
```
