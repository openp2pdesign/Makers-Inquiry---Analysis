# Q016 - Che tipo di contratto hai?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Get the distribution of contract type
contract = data["Q016"].value_counts(dropna=False)
print "Data:"
print contract
print ""
print "Data %:"
print data["Q016"].value_counts(normalize=True,dropna=False) * 100

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Tipi di contratto')
plt.ylabel('Persone')
my_colors = seaborn.color_palette("husl", len(contract)) # Set color palette
contract.plot(kind="bar",title="Che tipo di contratto hai?",color=my_colors)
plt.savefig("pdf/Q016-MakerTipiContratto01.pdf") # Save pdf
```

```python
# Get "other" data
contract_other = data["Q016[other]"].str.lower().value_counts()
print "Data:"
print contract_other
print ""
print "Data %:"
print data["Q016[other]"].str.lower().value_counts(normalize=True) * 100

# Plot bar
plt.figure(figsize=(8,6))
plt.title('Che tipo di contratto hai? Altro')
plt.xticks(range(len(contract_other.index)),contract_other.index,rotation=90)
plt.xlabel('Tipi di contratto')
plt.ylabel('Persone')
ind = np.arange(len(contract_other))   # the x locations for the groups
width = 0.35                         # the width of the bars

my_colors = seaborn.color_palette("husl", len(contract_other)) # Set color palette
rect1 = plt.bar(ind,contract_other,width,color=my_colors,align='center')
plt.savefig("pdf/Q016-MakerTipiContratto02.pdf") # Save pdf
```

```python
# Filter and integrate "other" data with data manually
contract["Collaborazione con/senza Partita IVA (Freelance)"] += 9
contract.ix["Titolare azienda"] = 9
contract.ix["Volontariato"] = 1
contract.ix["Assegno di ricerca"] = 1
contract.ix["Apprendistato"] = 1
contract.ix["Non specificato"] = 1
contract = contract.drop("Altro")
print contract
```

```python
# Plot updated bar
plt.figure(figsize=(8,6))
plt.title('Che tipo di contratto hai?')
plt.xticks(range(len(contract.index)),contract.index,rotation=90)
plt.xlabel('Tipi di contratto')
plt.ylabel('Persone')
ind = np.arange(len(contract))   # the x locations for the groups
width = 0.35                         # the width of the bars

my_colors = seaborn.color_palette("husl", len(contract)) # Set color palette
rect1 = plt.bar(ind,contract,width,color=my_colors,align='center')
plt.savefig("pdf/Q016-MakerTipiContratto01.pdf") # Save pdf
```
