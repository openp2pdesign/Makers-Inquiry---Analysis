# Q014 - Qual è la tua attività principale?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Get the distribution of activity type
activity_type = data["Q014"].value_counts(dropna=False)
print "Data:"
print activity_type
print ""
print "Data %:"
print data["Q014"].value_counts(normalize=True,dropna=False) * 100

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Tipo di Attivitá')
plt.ylabel('Persone')
my_colors = seaborn.color_palette("husl", len(activity_type)) # Set color palette
activity_type.plot(kind="bar",title=u"Qual è la tua attività principale?",color=my_colors)
plt.savefig(u"pdf/Q014-MakerTipoAttivitá01.pdf") # Save pdf
```

```python
# Gather makers employed

# Get sum of employed people
employed = activity_type.ix["Sono un lavoratore dipendente (piccola/media impresa)"] + activity_type.ix["Sono un lavoratore dipendente (multinazionale)"] + activity_type.ix["Sono un lavoratore dipendente (grande impresa)"] + activity_type.ix["Sono un lavoratore dipendente (microimpresa)"]

# Remove other types of employment
activity_type = activity_type.drop("Sono un lavoratore dipendente (piccola/media impresa)")
activity_type = activity_type.drop("Sono un lavoratore dipendente (multinazionale)")
activity_type = activity_type.drop("Sono un lavoratore dipendente (grande impresa)")
activity_type = activity_type.drop("Sono un lavoratore dipendente (microimpresa)")

# Add sum of employment
activity_type.ix["Sono un dipendente"] = employed
print activity_type
```

```python
# Plot the update data
plt.figure(figsize=(8,6))
plt.xlabel(u'Tipo di Attivitá')
plt.ylabel('Persone')
my_colors = seaborn.color_palette("husl", len(activity_type)) # Set color palette
activity_type.plot(kind="bar",title=u"Qual è la tua attività principale?",color=my_colors)
plt.savefig(u"pdf/Q014-MakerTipoAttivitá03.pdf") # Save pdf
```
