# Q012 - Qual è il tuo reddito annuale (espresso in fasce)?

```python
# -*- coding: UTF-8 -*-

# Render our plots inline
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier, overridden by seaborn
pd.set_option('display.max_columns', None) # Display all the columns
plt.rcParams['font.family'] = 'sans-serif' # Sans Serif fonts for all the graphs

# Reference for color palettes: http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html
```

```python
# Load csv file first
data = pd.read_csv("data/results-makers-40.csv", encoding="utf-8")
```

```python
# Check data
#data[0:4] # Equals to data.head()
```

```python
# Get the distribution of way of living
income = data["Q012"].value_counts(dropna=False)
print "Data:"
print income
print ""
print "Data %:"
print data["Q012"].value_counts(normalize=True,dropna=False) * 100

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Reddito')
plt.ylabel('Persone')
my_colors = seaborn.color_palette("husl", len(income)) # Set color palette
income.plot(kind="bar",title="Maker: Reddito annuale",color=my_colors)
plt.savefig("pdf/Q012-MakerReddito01.pdf") # Save pdf
```

```python
# Order of the choices
income_order = ["Non ho reddito",u"Meno di 10.000 €",u"Meno di 25.000 €",u"Meno di 50.000 €",u"Meno di 100.000 €",u"Oltre 100.000 €"]

income2 = income.reindex(income_order)

# Get the distribution of way of living, reindexed
print "Data:"
print income2
print ""
print "Data %:"
income2_normalized = data["Q012"].value_counts(normalize=True) * 100
print income2_normalized.reindex(income_order)

# Plot the data
plt.figure(figsize=(8,6))
plt.xlabel(u'Reddito')
plt.ylabel('Persone')
my_colors = seaborn.color_palette("husl", len(income2)) # Set color palette
income2.plot(kind="bar",title="Maker: Reddito annuale",color=my_colors)
plt.savefig("pdf/Q012-MakerReddito02.pdf") # Save pdf
```
